```python
import subprocess
from PIL import Image
import os

pdf_files = [
    "C101_025_blended_w0.1_0.9.pdf",
    "C101_025_blended_w0.2_0.8.pdf",
    "C101_025_blended_w0.3_0.7.pdf",
    "C101_025_blended_w0.4_0.6.pdf",
    "C101_025_blended_w0.5_0.5.pdf",
    "C101_025_blended_w0.6_0.4.pdf",
    "C101_025_blended_w0.7_0.3.pdf",
    "C101_025_blended_w0.8_0.2.pdf",
    "C101_025_blended_w0.9_0.1.pdf",
    "C101_025_hierarchical.pdf"
]

image_paths = []
for i, pdf_file in enumerate(pdf_files):
    output_prefix = f"temp_grid_{i}"
    # Keeping the high-quality 600 DPI setting
    subprocess.run(['pdftoppm', '-r', '600', '-png', '-singlefile', pdf_file, output_prefix])
    image_paths.append(f"{output_prefix}.png")

images = [Image.open(p) for p in image_paths]

# Dimensions
w, h = images[0].size

# Define layouts as (columns, rows)
# 5 collages x 2 slots = exactly 10 slots for your 10 images
layouts = [
    (1, 2),  # Collage 1
    (1, 2),  # Collage 2
    (1, 2),  # Collage 3
    (1, 2),  # Collage 4
    (1, 2)   # Collage 5
]

image_index = 0

for i, (cols, rows) in enumerate(layouts):
    if image_index >= len(images):
        break  # We've processed all images
        
    # Create a blank white canvas for the current layout
    collage = Image.new('RGB', (cols * w, rows * h), (255, 255, 255))
    capacity = cols * rows
    
    # Fill the current collage with the available images
    for pos in range(capacity):
        if image_index >= len(images):
            break  # Stop if we run out of images before filling the grid
            
        r = pos // cols
        c = pos % cols
        collage.paste(images[image_index], (c * w, r * h))
        image_index += 1
        
    collage_name = f"collage_C101_025_part{i+1}.png"
    collage.save(collage_name)
    print(f"Collage {i+1} saved as {collage_name} (Layout: {cols}x{rows})")

# Cleanup temporary PNGs
for p in image_paths:
    if os.path.exists(p):
        os.remove(p)

print("All collages generated and temporary files cleaned up.")
```
