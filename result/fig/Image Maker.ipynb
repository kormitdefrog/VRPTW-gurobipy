```python
import subprocess
from PIL import Image
import os

pdf_files = [
    "R101_050_blended_w0.1_0.9.pdf",
    "R101_050_blended_w0.2_0.8.pdf",
    "R101_050_blended_w0.3_0.7.pdf",
    "R101_050_blended_w0.4_0.6.pdf",
    "R101_050_blended_w0.5_0.5.pdf",
    "R101_050_blended_w0.6_0.4.pdf",
    "R101_050_blended_w0.7_0.3.pdf",
    "R101_050_blended_w0.8_0.2.pdf",
    "R101_050_blended_w0.9_0.1.pdf",
    "R101_050_hierarchical.pdf"
]

image_paths = []
for i, pdf_file in enumerate(pdf_files):
    output_prefix = f"temp_grid_{i}"
    subprocess.run(['pdftoppm', '-png', '-singlefile', pdf_file, output_prefix])
    image_paths.append(f"{output_prefix}.png")

images = [Image.open(p) for p in image_paths]

# Dimensions
w, h = images[0].size
cols = 3
rows = 4 # 3 rows for first 9, 1 row for the 10th

collage = Image.new('RGB', (cols * w, rows * h), (255, 255, 255))

# Paste first 9 images in 3x3 grid
for i in range(9):
    r = i // cols
    c = i % cols
    collage.paste(images[i], (c * w, r * h))

# Paste the 10th image in the middle of the 4th row
collage.paste(images[9], (1 * w, 3 * h))

collage_name = "collage_R101_050.png"
collage.save(collage_name)

# Cleanup
for p in image_paths:
    if os.path.exists(p):
        os.remove(p)

print(f"Collage saved as {collage_name}")
```
