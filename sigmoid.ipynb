```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the modified sigmoid function
def modified_sigmoid(x, a=1, b=0, lb=0.2, ub=0.8):
    """
    Equation: y = lb + (ub - lb) * (1 / (1 + exp(-a*(x-b))))
    a: controls the steepness
    b: shifts the curve horizontally (inflection point)
    lb: lower bound
    ub: upper bound
    """
    return lb + (ub - lb) * (1 / (1 + np.exp(-a * (x - b))))

# 2. Generate an array of x values
x = np.linspace(0, 5, 200)

# 3. Set parameters
a_values = [1, 2, 3] # Array of a_values to iterate through
b_val = 4.0 

# 4. Create a figure with 1 row and 3 columns of subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5)) 

# 5. Iterate through the a_values and the corresponding subplot axes
for i, a_val in enumerate(a_values):
    ax = axes[i] # Select the current subplot
    y = modified_sigmoid(x, a=a_val, b=b_val)
    
    # Create the plot on the current subplot
    ax.plot(x, y, color='purple', linewidth=2, label=f'$P_k$ (a={a_val}, b={b_val})')
    
    # Add formatting and new asymptotes (Note: using ax.set_title instead of plt.title)
    ax.set_title(f'Sigmoid Function (Slope = {a_val})', fontsize=14)
    ax.set_xlabel('Rating', fontsize=12)
    ax.set_ylabel('$P_k$', fontsize=12)
    
    # Draw the standard axes
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    
    # Draw the new upper and lower bounds based on the formula
    ax.axhline(0.8, color='red', linestyle='--', alpha=0.5, label='y = 0.8 (Upper Bound)')
    ax.axhline(0.2, color='green', linestyle='--', alpha=0.5, label='y = 0.2 (Lower Bound)')
    
    ax.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
    ax.legend(loc='lower right', fontsize=9)

# Adjust layout to prevent overlapping text
plt.tight_layout()

# 6. Display the plot
plt.show()
```
